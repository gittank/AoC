```python
import urllib.request

def gather_input_data(url, sessionId, transform=lambda x: str(x, "utf-8").strip('\n')):
    request = urllib.request.Request(url)
    request.add_header("cookie", "session={}".format(sessionId)) # Source the data directly from AoC

    values = []
    with urllib.request.urlopen(request) as data:
        for line in data:
            values.append(transform(line))

    return values

def get_data(day, year=2023):
    with open('sessionID') as f:
        sessionId = f.readlines()[0].strip()
    url = "https://adventofcode.com/%d/day/%d/input"%(year,day)
    data = gather_input_data(url, sessionId)
    return data
    

from collections import deque
import re

def move_beam(state):
    x,y,direction = state
    x += dir_to_xy[direction][0]
    y += dir_to_xy[direction][1]
    return (x,y,direction)
    
def new_direction(state):
    change_direction ={'/' :{RIGHT:[UP],      UP:[RIGHT],       DOWN:[LEFT],     LEFT:[DOWN]},
                       '\\':{RIGHT:[DOWN],    UP:[LEFT],        DOWN:[RIGHT],    LEFT:[UP]},
                       '.' :{RIGHT:[RIGHT],   UP:[UP],          LEFT:[LEFT],     DOWN:[DOWN]},
                       '-' :{RIGHT:[RIGHT],   UP:[RIGHT, LEFT], LEFT:[LEFT],     DOWN:[RIGHT, LEFT]},
                       '|' :{RIGHT:[UP,DOWN], UP:[UP],          LEFT:[UP, DOWN], DOWN:[DOWN]}
                      }

    x,y,direction = state
    tile = data[x][y]
    return change_direction[tile][direction]        

def energized_count(visited):
    out = set()
    for state in visited:
        x,y,_ = state
        out.add((x,y))
    return(len(out))

def go(explore):
    right_edge = len(data[0])
    left_edge = -1
    top_edge = -1
    bottom_edge = len(data)

    while explore:
        here = explore.popleft()
        new = move_beam(here)
    
        if not (left_edge < new[1] < right_edge) or not (top_edge < new[0] < bottom_edge):
            continue
        if new in visited:
            continue
        visited.add(new)
    
        for directions in new_direction(new):
            explore.extend([(new[0],new[1],directions)])  
    return visited
```

```python
data = get_data(16, 2023)


#    0
# 1     2
#    3

dir_to_xy = [(-1,0), (0,-1), (0,1), (1,0)]
UP = 0
LEFT = 1
RIGHT = 2
DOWN = 3



visited = set()
start = (0,-1,RIGHT)


explore = deque([start])
visited = go(explore)
energized_count(visited)


scores = []
for ii in range(len(data)):
    visited=set()
    start = (ii,-1,RIGHT)
    explore = deque([start])
    visited = go(explore)
    scores.append(energized_count(visited))

    visited=set()
    start = (ii,len(data),LEFT)
    explore = deque([start])
    visited = go(explore)
    scores.append(energized_count(visited))
    
    visited=set()
    start = (-1,ii,DOWN)
    explore = deque([start])
    visited = go(explore)
    scores.append(energized_count(visited))
    
    visited=set()
    start = (len(data),ii,UP)
    explore = deque([start])
    visited = go(explore)
    scores.append(energized_count(visited))
  
print(scores[0], max(scores))
```
