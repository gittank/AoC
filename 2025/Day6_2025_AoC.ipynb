```python
import urllib.request

def gather_input_data(url, sessionId, transform=lambda x: str(x, "utf-8").strip('\n')):
    request = urllib.request.Request(url)
    request.add_header("cookie", "session={}".format(sessionId)) # Source the data directly from AoC

    values = []
    with urllib.request.urlopen(request) as data:
        for line in data:
            values.append(transform(line))

    return values

def get_data(day, year=2025):
    with open('sessionID') as f:
        sessionId = f.readlines()[0]
    url = "https://adventofcode.com/%d/day/%d/input"%(year,day)
    data = gather_input_data(url, sessionId)
    return data
```

```python
data = [\
'123 328  51 64 ', 
' 45 64  387 23 ', 
'  6 98  215 314',
'*   +     *    + '] 
data = get_data(6)

```

```python
import re
numbers = [list(map(int, re.findall('-?[0-9]+', val))) for val in data[:-1]]
transposed_numbers = [list(row) for row in zip(*numbers)]
operators = re.findall('[+,*]', data[-1])

```

```python
total = 0
for ii, val in enumerate(transposed_numbers):
    op = operators[ii]
    if op == '+':
        sub_total = sum(val)
    else:
        sub_total = 1
        for num in val:
            sub_total *= num
    total += sub_total
print(total)
```

```python
op_cnt = 0
ii = 0
total = 0

while(ii<len(data[0])):
    all_space = False
    if operators[op_cnt] == '+':
        sub_total = 0
    else:
        sub_total = 1
        
    while (not all_space) and (ii<len(data[0])):
        number = ''
        for row in data[:-1]:
            number+=row[ii]
        if number.isspace():
            all_space = True
            op_cnt += 1
        else:
            number = int(number)
            if operators[op_cnt] == '+':
                sub_total += number
            else:
                sub_total *= number
        ii += 1
    total = total + sub_total
print(total)
```
