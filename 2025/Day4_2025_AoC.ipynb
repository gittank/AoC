```python
import urllib.request

def gather_input_data(url, sessionId, transform=lambda x: str(x, "utf-8").strip('\n')):
    request = urllib.request.Request(url)
    request.add_header("cookie", "session={}".format(sessionId)) # Source the data directly from AoC

    values = []
    with urllib.request.urlopen(request) as data:
        for line in data:
            values.append(transform(line))

    return values

def get_data(day, year=2025):
    with open('sessionID') as f:
        sessionId = f.readlines()[0]
    url = "https://adventofcode.com/%d/day/%d/input"%(year,day)
    data = gather_input_data(url, sessionId)
    return data

```

```python
def get_8_neighbors(grid, row, col):
    """
    Returns a list of the 8 neighbors of a character at a given (row, col) in a grid.

    Args:
        grid: A list of strings representing the grid.
        row: The row index of the character.
        col: The column index of the character.

    Returns:
        A list of characters representing the neighbors.
    """
    neighbors = []
    num_rows = len(grid)
    num_cols = len(grid[0]) if num_rows > 0 else 0

    # Define the 8 possible relative offsets for neighbors
    # (dr, dc) represents (delta row, delta column)
    directions = [
        (-1, -1), (-1, 0), (-1, 1),  # Top row
        (0, -1),           (0, 1),   # Middle row (excluding self)
        (1, -1),  (1, 0),  (1, 1)    # Bottom row
    ]

    for dr, dc in directions:
        neighbor_row, neighbor_col = row + dr, col + dc

        # Check if the neighbor's coordinates are within the grid boundaries
        if 0 <= neighbor_row < num_rows and 0 <= neighbor_col < num_cols:
            neighbors.append(grid[neighbor_row][neighbor_col])

    return neighbors

def fewer_than_4(span):
    return ''.join(span).count('@') < 4

def remove_rolls(diagram, remove_list):
    for x,y in remove_list:
        diagram[x] = diagram[x][:y] + '*' + diagram[x][y+1:]
    return diagram
```

```python
part1 = 0
for row in range(len(data)):
    for col in range(len(data[0])):
        if data[row][col] == '@':
            #print(row, col, get_8_neighbors(data, row, col), fewer_than_4(get_8_neighbors(data, row, col)))
            part1 += fewer_than_4(get_8_neighbors(data, row, col))

print(part1)
```

```python
part2 = 0
found_something = True
to_be_removed =[]

while found_something:
    for row in range(len(data)):
        for col in range(len(data[0])):
            if data[row][col] == '@':
                if fewer_than_4(get_8_neighbors(data, row, col)):
                    part2 += 1
                    to_be_removed.append((row, col))
    found_something = len(to_be_removed) != 0
    data = remove_rolls(data, to_be_removed)
    to_be_removed = []
print(part2)
```
