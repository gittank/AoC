```python
import urllib.request

def gather_input_data(url, sessionId, transform=lambda x: str(x, "utf-8").strip('\n')):
    request = urllib.request.Request(url)
    request.add_header("cookie", "session={}".format(sessionId)) # Source the data directly from AoC

    values = []
    with urllib.request.urlopen(request) as data:
        for line in data:
            values.append(transform(line))

    return values

def get_data(day, year=2025):
    sessionId = '53616c7465645f5f702b530d8b2e71556ed5068d0e24fec5cb9d6ca1f98bfb9891ddf8b524837fd60a53e9788aa179280dc66452bd70e26b9fbe32205ffde952'
    # with open('sessionID') as f:
    #     sessionId = f.readlines()[0]
    url = "https://adventofcode.com/%d/day/%d/input"%(year,day)
    data = gather_input_data(url, sessionId)
    return data
```

```python
def rotate_positive(now, delta):
    step = 100 - now
    if delta < step:
        add_count = 0
        new = now + delta
    else:
        delta = delta - step
        add_count = 1 + (delta // 100)
        new = (delta % 100)
        if new == 100:
            new = 0
    return new, add_count

def rotate_negative(now, delta):
    if delta < now:
        add_count = 0
        new = now - delta
    else:
        delta = delta - now
        if now == 0:
            add_count = (delta // 100)
        else:
            add_count = 1 + (delta // 100)

        new = 100 - (delta % 100)
        if new == 100:
            new = 0
    return new, add_count
```
